### Подбор гиперпараметров

Создадим функцию для автоматичесного подбора параметров моделей спомощью GridSearchCV.
Будем подавать обучающую выборку для обучения моделей, а также параметры, которые GridSearchCV обработает и вернет нам наилучший набор параметров модели для обучающих данных.

```python
def grid_search(X, y, model, parameters):#функция для автоматического подбора параметров
    model_cv = GridSearchCV(model, parameters, cv=4,  scoring='f1')#спомощью GridSearchCV подбираем параметры
    model_cv.fit(X, y)#обучаем модель
    return model_cv.best_params_#возвращаем наилучшие параметры
```

Найдем наилучшие параметры для модели дерево решений.

```python
tree_param = {'criterion': ['entropy', 'gini'], 'max_depth': range (1,16, 2)}#Отдельно создаем словарик, в который вписываем параметры, которые будем прогонять GridSearch’ем. 
grid_search(features_train, target_train, DecisionTreeClassifier(), tree_param)# передаем функции grid_search параметры
```

Найдем наилучшие параметры для модели случайный лес.

```python
#Отдельно создаем словарик, в который вписываем параметры, которые будем прогонять GridSearch’ем. Для примера будем использовать следующие параметры:
rfc_param = { 'n_estimators': range (10, 51, 10),#n_estimators – число деревьев в лесу. Оно будет изменяться от 10 до 50 с шагом 10
              'max_depth': range (1,13, 2),#max_depth – глубина дерева. Она будет изменяться от 1 до 12 с шагом в 2
              'min_samples_leaf': range (1,6),#min_samples_leaf – минимальное число образцов в листах. Оно будет изменяться от 1 до 7
              'min_samples_split': range (2,8,2)#min_samples_leaf – минимальное число образцов для сплита. Оно будет изменяться от 2 до 9.
           }

              
grid_search(features_train, target_train, RandomForestClassifier(), rfc_param)# передаем функции grid_search параметры
```

Найдем наилучшие параметры для модели логистической регрессии.

```python
solver_options = ['newton-cg', 'lbfgs', 'sag']
multi_class_options = ['ovr', 'multinomial']
C = [0.1, 1, 10]

log_param = dict(solver = solver_options, multi_class = multi_class_options, C=C)#Отдельно создаем словарик, в который вписываем параметры, которые будем прогонять GridSearch’ем. 
              
grid_search(features_train, target_train, LogisticRegression(), log_param)# передаем функции grid_search параметры
```

### Выбор модели

Перейдем к выбору лучшей модели для решения нашей задачи.
Создадим цикл, в котором для каждой модели посчитаем метрики: 
 - аккуратность - доля правильных ответов алгоритма, 
 - F1-меру - гармоническое среднее между точностью и полнотой,
 - ROC AUC - площадь под ROC-кривой.

```python
def model_study(dtc, rfc, log, feat_train, targ_train):
    models = [dtc, rfc, log]

    results = []

    for model_i in models:
        model_i.fit(feat_train, targ_train)
        predictions_valid = model_i.predict(features_valid)
        probabilities_valid = model_i.predict_proba(features_valid)
        predictions_test = model_i.predict(features_test)
        probabilities_test = model_i.predict_proba(features_test)
        predictions_test = model_i.predict(features_test)
    
        accuracy = accuracy_score(target_valid, predictions_valid)
        accuracy_test = accuracy_score(target_test, predictions_test)
        f1 = f1_score(target_valid, predictions_valid)
        f1_test = f1_score(target_test, predictions_test)
        roc_auc = roc_auc_score(target_valid, probabilities_valid[:,1])
        roc_auc_test = roc_auc_score(target_test, probabilities_test[:,1])
        diff =  accuracy - accuracy_test
    
        results.append({'model': model_i.__class__.__name__, 
                               'accuracy': accuracy, 
                              'accuracy_test': accuracy_test,
                               'retraining': diff,
                               'f1': f1, 
                               'f1_test': f1_test, 
                               'ROC AUC': roc_auc,                            
                               'ROC AUC_test': roc_auc_test,})
        
    
    df_result=pd.DataFrame(results)
    return df_result
```

```python
model_study(DecisionTreeClassifier(random_state=12345, criterion = 'gini', max_depth = 5), 
            RandomForestClassifier(random_state=12345, max_depth = 11, min_samples_leaf = 2, min_samples_split = 6, n_estimators = 50),
                 LogisticRegression(random_state=12345, C= 10, multi_class= 'ovr', solver= 'newton-cg'), features_train, target_train)
```

Метрика F1 показывает баланс между precision (точностью) и recall (полнотою). Мы используем его при неравномерном распределении классов, так как точность и полнота могут дать неверные результаты!

Таким образом, мы используем показатель F1 в качестве индикатора сравнения между точностью и полнотою.

ROC AUC сравнивает истинно-положительный уровень с ложно-положительным уровнем.

Таким образом, чем больше ROC AUC, тем больше различие между истинно-позитивным и истинно-негативавными уровнями.

Следовательно, когда у вас есть дисбаланс данных между положительными и отрицательными выборками, мы всегда должны использовать F1-показатель, потому что ROC усредняется по всем возможным порогам.

## Борьба с дисбалансом

В рассматриваемой нами задаче классификации предпочтительно, чтобы обучающие примеры были относительно равномерно
распределены среди классов, однако в нашем случае они не сбалансированы, что приводит к проблеме качественной оценки работы классификатора. Классическая метрика аккуратности (accuracy) может выдавать высокое значение качества, при этом в реальности модель работает хуже прогноза, всегда определяющего один класс.

Придадим объектам редкого класса больший вес (англ. class_weight, «вес класса»).

### Взвешивание классов

В библиотеке sklearn есть аргумент class_weight придадим ему значение balanced  (англ. «сбалансированный»), алгоритм посчитает, во сколько раз класс «0» встречается чаще класса «1». Бóльший вес будет у редкого класса.

```python
model_study(DecisionTreeClassifier(random_state=12345, criterion = 'gini', max_depth = 5,  class_weight='balanced'), 
            RandomForestClassifier(random_state=12345, max_depth = 11, min_samples_leaf = 2, min_samples_split = 6, n_estimators = 50,  class_weight='balanced'),
                 LogisticRegression(random_state=12345, C= 10, multi_class= 'ovr', solver= 'newton-cg',  class_weight='balanced'), 
            features_train, target_train)
```

Проверим, как другие методы борьбы с дисбалансом повлияют на метрики.

### Увеличение выборки

Сделаем объекты редкого класса не такими редкими в данных. Пусть они повторяются по несколько раз, чтобы их легче запомнить.
Такая техника называется upsampling (от англ. up, «вверх»; sampling, «выборка»).
Преобразование проходит в несколько этапов:
*	Разделим обучающую выборку на отрицательные и положительные объекты;
*	Скопируем несколько раз положительные объекты;
*	С учётом полученных данных создаим новую обучающую выборку;
*	Перемешаем данные: идущие друг за другом одинаковые вопросы не помогут обучению.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 3)
```

```python
model_study(DecisionTreeClassifier(random_state=12345, criterion = 'gini', max_depth = 5), 
            RandomForestClassifier(random_state=12345, max_depth = 11, min_samples_leaf = 2, min_samples_split = 6, n_estimators = 50),
                 LogisticRegression(random_state=12345, C= 10, multi_class= 'ovr', solver= 'newton-cg'), features_upsampled, target_upsampled)
```

### Уменьшение выборки

По аналогии с увеличением выборки, проделаем туже операцию, но в обратную сторону. Уберём часть неважных признаков. Это можно сделать техникой downsampling (от англ. down, «вниз»; sampling, «выборка»).
Преобразование проходит в несколько этапов:
*	Разделить обучающую выборку на отрицательные и положительные объекты;
*	Случайным образом отбросить часть из отрицательных объектов;
*	С учётом полученных данных создать новую обучающую выборку;
*	Перемешать данные. Положительные не должны идти следом за отрицательными: алгоритмам будет сложнее обучаться.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.35)
```

```python
model_study(DecisionTreeClassifier(random_state=12345, criterion = 'gini', max_depth = 5), 
            RandomForestClassifier(random_state=12345, max_depth = 11, min_samples_leaf = 2, min_samples_split = 6, n_estimators = 50),
                 LogisticRegression(random_state=12345, C= 10, multi_class= 'ovr', solver= 'newton-cg'), features_downsampled, target_downsampled)
```

- логистическая регрессия, показала лучший результат F1 = 0,49 на тестовой выборке при уменьшении выборки. Коэффициент уменьшения выборки равен 0,35.
- случайный лес при методе борьбы с дисбалансом - увеличением выборки. Коэффициент увеличение выборки равен 3. Для данной модели метрики равны:

* most_frequent - Если вы проанализировали большое количество этих объектов и обнаружили, что 90% содержат целевое свойство, то угадывание того, что каждый будущий экземпляр объекта обладает целевым свойством, дает 90% -ную вероятность правильного угадывания.
* stratified - этот классификатор будет предсказывать, что существует 90% вероятности того, что каждый объект, с которым он сталкивается, обладает целевым свойством
* most_frequent - предположил бы, что все будущие объекты обладают целевым свойством.

```python
def DC(strateg):
    dummy_clf = DummyClassifier(strategy=strateg)
    dummy_clf.fit(features_train, target_train)
    DummyClassifier(strategy=strateg)
    dummy_clf.predict(features_train)
    print('аccuracy на фиктивном классификаторе при стратегии', strateg,  'составила:', dummy_clf.score(features_train, target_train))
```
